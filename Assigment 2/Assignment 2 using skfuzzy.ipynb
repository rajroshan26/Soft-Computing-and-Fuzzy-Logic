```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,7)
import skfuzzy
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
from skfuzzy.membership import trimf,gauss2mf,gaussmf,gbellmf,sigmf
```

```python
plt.rcParams["figure.figsize"] = (20,20)
x = np.arange(0,20)
mfx = fuzz.trimf(x, [5, 10, 15])
a = np.arange(0,30)
print ("X is",+x)
print ("mf of X is ",mfx)
print("Universal Set Of X is",+a)
plt.subplot(311)
plt.plot(x,mfx,"r")
```

```python
mfx=fuzz.gauss2mf(x,5,2,12,3)
print ("X is",+x)
print ("mf of X is ",mfx)
print("Universal Set Of X is",+a)
plt.subplot(311)
plt.plot(x,mfx,"r")
```

```python
mfx=fuzz.gaussmf(x,10,2.9)
print ("X is",+x)
print ("mf of X is ",mfx)
print("Universal Set Of X is",+a)
plt.subplot(311)
plt.plot(x,mfx,"r")
```

```python
mfx=fuzz.gbellmf(x,10,5,0.5)
print ("X is",+x)
print ("mf of X is ",mfx)
print("Universal Set Of X is",+a)
plt.subplot(311)
plt.plot(x,mfx,"r")
```
