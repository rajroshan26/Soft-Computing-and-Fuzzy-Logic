```python
import skfuzzy
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
from skfuzzy.membership import trimf
```

```python
plt.rcParams["figure.figsize"] = (20,20)
x = np.arange(0,20)
mfx = fuzz.trimf(x, [5, 10, 15])
a = np.arange(0,30)
print ("X is",+x)
print ("mf of X is ",mfx)
print("Universal Set Of X is",+a)
plt.subplot(311)
plt.plot(x,mfx,"r")

y = np.arange(10,30)
mfy = fuzz.trimf(y, [11, 15, 21])
b = np.arange(30)
print ("Y is",+y)
print ("mf of Y is ",mfy)
print("Universal Set Of Y is",+b)
plt.subplot(312)
plt.plot(y,mfy,"g")

plt.subplot(313)
plt.plot(x,mfx,"r")
plt.plot(y,mfy,"g")
```

```python
z,mfz = skfuzzy.fuzzymath.fuzzy_add(x,mfx,y,mfy)
print(z)
print(mfz)

#     Add fuzzy set ``a`` to fuzzy set ``b``.

#     Parameters
#     ----------
#     x : 1d array, length N
#         Universe variable for fuzzy set ``a``.
#     a : 1d array, length N
#         Fuzzy set for universe ``x``.
#     y : 1d array, length M
#         Universe variable for fuzzy set ``b``.
#     b : 1d array, length M
#         Fuzzy set for universe ``y``.

#     Returns
#     -------
#     z : 1d array
#         Output variable.
#     mfz : 1d array
#         Fuzzy membership set for variable ``z``.
plt.rcParams["figure.figsize"] = (20,7)
plt.plot(z,mfz)
```

```python
q,mfq = skfuzzy.fuzzymath.fuzzy_sub(x,mfx,y,mfy)
print(q)
print(mfq)
plt.plot(q,mfq)
```

```python
skfuzzy.fuzzymath.cartprod(x,y)

# Cartesian product of two fuzzy membership vectors. Uses ``min()``.

#     Parameters
#     ----------
#     x : 1D array or iterable
#         First fuzzy membership vector, of length M.
#     y : 1D array or iterable
#         Second fuzzy membership vector, of length N.

#     Returns
#     -------
#     z : 2D array
#         Cartesian product of ``x`` and ``y``, of shape (M, N).
```

```python
r,mfr = skfuzzy.fuzzymath.fuzzy_or(x, mfx, y, mfy)
print(r)
print(mfr)
plt.plot(r,mfr)
```

```python
p,mfp = skfuzzy.fuzzymath.fuzzy_and(x, mfx, y, mfy)
print(p)
print(mfp)
plt.plot(p,mfp)
```
